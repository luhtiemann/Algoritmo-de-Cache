```python
"""import random
import time
from google.colab import drive
drive.mount('/content/drive',force_remount=True)

%cd /content/drive/MyDrive/Cash/

# Criando lista com os 100 textos
texto = []
for i in range(1, 101):
  with open(f"{i}.txt", "r") as f:
    texto.append(f.read())"""

import random
import time
import os

# Clonando o repositório do GitHub
!git clone https://github.com/luhtiemann/cache.git

# Mudando para o diretório do repositório clonado
os.chdir("cache")

# Criando lista com os 100 textos
texto = []
for i in range(1, 101):
  with open(f"{i}.txt", "r") as f:
    texto.append(f.read())

# Cache
cache = {}
cache_size = 10

# Peso de 33% para os numeros entre 30 a 40
def random_number():
    if random.randint(0, 2) == 0:
        return random.randint(30, 40)
    else:
        return random.randint(1, 29) if random.randint(0, 1) == 0 else random.randint(41, 99)

# Implementação do algoritmo de cache FIFO
def fifo(texto):
    if len(cache) == cache_size:
        cache.pop(list(cache.keys())[0])
    cache[texto] = time.time()

# Implementação do algoritmo de cache LRU
def lru(texto):
    if len(cache) == cache_size:
        cache.pop(min(cache, key=cache.get))
    cache[texto] = time.time()

# Implementação do algoritmo de cache LFU
def lfu(texto):
    if len(cache) == cache_size:
        least_frequent = min(cache.items(), key=lambda x: x[1])[0]
        cache.pop(least_frequent)
    cache[texto] = time.time()

# Simulação do acesso de usuários aleatórios
def simulacao(algoritmo):
    tempo_total = 0
    cache_miss = 0
    for i in range(200):
        # Escolha aleatória do texto
        temp = random_number()
        # Verifica se o texto está no cache
        if temp in cache:
            # Cache hit
            print(f"Arquivo {temp}.txt encontrado no cache")
            tempo_inicio = time.time()
            with open(f"{temp}.txt", "r") as f:
                f.read()
            tempo_total += time.time() - tempo_inicio
        else:
            # Cache miss
            cache_miss += 1
            print(f"Arquivo {temp}.txt não encontrado no cache")
            tempo_inicio = time.time()
            with open(f"{temp}.txt", "r") as f:
                f.read()
            tempo_total += time.time() - tempo_inicio
            # Adiciona o arquivo ao cache
            algoritmo(temp)

    # Cálculo do tempo médio e número de cache miss
    tempo_medio = tempo_total / 200
    cash_hit = 200-cache_miss

    return (tempo_medio, cache_miss, cash_hit)



# Pergunta ao usuário se ele quer simular ou sair do programa
num_texto = int(input("Digite o número do texto que deseja ler (0 para sair, -1 para simulação): "))

# Loop principal do programa
while num_texto == -1:
    # Verifica se o usuário deseja sair
    if num_texto == 0:
        print("Encerrando o programa...")
        break
    # Simulação com escolha aleatória de texto


    # Simulação da leitura de todos os arquivos pelos três usuários
    usuarios = ["Usuario 1", "Usuario 2", "Usuario 3"]
    algoritmos = {"FIFO": fifo, "LRU": lru, "LFU": lfu}

    with open("relatorio.txt", "w") as f:
      f.write("Relatório de Simulação\n\n")
      for usuario in usuarios:
        f.write(f"{usuario}\n\n")
        for algoritmo_nome, algoritmo_func in algoritmos.items():
          f.write(f"{algoritmo_nome}\n")
          tempo_medio, cache_miss, cash_hit = simulacao(algoritmo_func)
          f.write(f"Tempo médio: {tempo_medio:.3f}s\n")
          f.write(f"Número de cache misses: {cache_miss}\n")
          f.write(f"Número de cache hit: {cash_hit}\n\n")

    with open('relatorio.txt', 'r') as f:
      print("\n\n", f.read())




    num_texto = int(input("\nDigite o número do texto que deseja ler (0 para sair, -1 para simulação): "))

    #pasta dos arquivos, inclusive o relatório foi direcionado para essa pasta
    #https://drive.google.com/drive/folders/15GE4M9QWm41umfl-KCdvOAQhM05i7niU?usp=sharing
```
