```python
import os

from google.colab import drive
drive.mount('/content/drive',force_remount=True)

file_list=[]

%cd /content/drive/MyDrive/

def split_file(filename, n_files, n_words):
    # Abrir o arquivo base para leitura
    with open(filename, 'r', encoding='utf8') as base_file:
        # Ler todas as palavras do arquivo base e armazená-las em uma lista
        words = base_file.read().split()

    # Calcular o número total de palavras por arquivo
    words_per_file = len(words) // n_files

    # Criar os arquivos de texto e escrever as palavras em cada um
    for i in range(n_files):
        # Criar o nome do arquivo com base no número atual do loop
        output_filename = f'{i+1}.txt'

        # Abrir o arquivo para escrita
        with open(output_filename, 'w',encoding='utf8') as output_file:
            # Escrever as palavras no arquivo atual
            start = i * words_per_file
            end = start + words_per_file if i != n_files - 1 else None
            output_file.write(' '.join(words[start:end]))

            # Adicionar a extensão do arquivo criado à lista de arquivos gerados
            file_list.append(output_filename)

    return file_list

# Exemplo de uso
base_filename = '/content/drive/MyDrive/Cash/base.txt'
n_files = 100
n_words = 1000
file_list = split_file(base_filename, n_files, n_words)
print(f'{n_files} arquivos criados: {", ".join(file_list)}')
```
